```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Plot for Part B
dataBh = pd.read_csv("AlgoProject2/dataBh.csv")
f, ax = plt.subplots(1, 1, figsize=(20, 10))
sb.lineplot(data=dataBh, x="h", y="Time")
ax.set_ylabel("Time (nanoseconds)")
plt.show()
```
